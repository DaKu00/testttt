```python
weight = input("당신의 몸무게는 몇 kg입니까?")
height = input("당신의 키는 몇 xm입니까?")
bmi = int(weight)/((int(height) * 0.01) ** 2)
if bmi < 18.5 :
    print("당신은 저체중입니다.")
elif 18.5 <= bmi and 22.9 > bmi :
    print("당신은 정상체중입니다.")
elif 22.9 <= bmi and 24.9 > bmi:
    print("당신은 과체중입니다.")
elif 24.9 <= bmi and 29.9 > bmi :
    print("당신은 경도비만입니다.")
else:
    print("당신은 고도비만입니다.")
```
