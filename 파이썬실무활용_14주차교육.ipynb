```python
import pandas as pd

folder = 'C:/OCU교육/data/'
excel_file = folder + '영업팀별판매현황.xlsx'
df = pd.read_excel(excel_file)
df
```

```python
excel_file_chart = folder + '영업팀별판매현황_차트.xlsx'
excel_writer = pd.ExcelWriter(excel_file_chart, engine = 'xlsxwriter')
worksheet_name = 'Sheet1'
df.to_excel(excel_writer, sheet_name = worksheet_name, index = False)
workbook = excel_writer.book
worksheet = excel_writer.sheets[worksheet_name]
```

```python
chart = workbook.add_chart({'type': 'line'})
columns_len = len(df.columns)
for k in range(columns_len - 1):
    start_row = 1
    start_column = k + 1
    end_row = len(df.index)
    end_column = k + 1
    chart.add_series({'values':[worksheet_name, start_row, start_column, end_row, end_column],
                     'categories': [worksheet_name, start_row, 0, end_row, 0],
                     'name': [worksheet_name, 0, k + 1],
                     'marker': {'type': 'circle'}})
```

```python
chart.set_title({'name': '팀별 판매 현황'})
chart.set_x_axis({'name': '월'})
chart.set_y_axis({'name': '판매량'})

worksheet.insert_chart(1, columns_len, chart, {'x_offset': 25, 'y_offset': 10})
excel_writer.save()
print('생성파일:', excel_file_chart)
```
